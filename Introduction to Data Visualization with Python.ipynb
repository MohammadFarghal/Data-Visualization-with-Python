###  we can't determine which is the faster because we use absolute values  

### we convert the numbers to growth by      


### 1-get the first value of gdp appear in eche 2 countries


### 2-divded every gpd value to the first one  to get  growth ratio 

### 3-  just multiply by 100

```python
# Get first recorded USA  GDP   
df_USA.gdpPercap.iloc[0]
#iloc integral location 
#we know DATAframe 
# consist of pandas series as columns so call first value of GDP  series
```

```python
df_USA.gdpPercap # the gdp series 
```

```python
df_USA.gdpPercap /df_USA.gdpPercap.iloc[0] #growth ratio
```

```python
df_USA.gdpPercap /df_USA.gdpPercap.iloc[0] * 100 # just multiply by 100
```

```python
USA_growth = df_USA.gdpPercap /df_USA.gdpPercap.iloc[0] * 100
China_growth =df_China.gdpPercap /df_China.gdpPercap.iloc[0]*100
```

```python
plt.plot(df_USA.year, USA_growth)
plt.plot(df_China.year, China_growth)
plt.legend(["USA","China"]) # legend take list of names
plt.title("USA  VS China GDP  ")
plt.ylabel("GDP Per capta")
plt.xlabel("years")
plt.show()
```

```python
obama=pd.read_csv("obama.csv", parse_dates=["year_month"])
```

```python
obama.head(5)
```

```python
plt.plot(obama.year_month,obama.approve_percent,"*",markersize =2
         ,alpha=.5)
```

```python
mean_df =obama.groupby("year_month").mean()
median_df =obama.groupby("year_month").median()
Q1 =obama.groupby("year_month").quantile(.25)
Q3 =obama.groupby("year_month").quantile(.75)
```

```python
plt.plot(mean_df.index , mean_df.approve_percent ,"red")
plt.plot(median_df.index , median_df.approve_percent, "gray")
plt.plot(Q3.index ,Q3.approve_percent,"blue")
plt.plot(Q1.index ,Q1.approve_percent,"green")
plt.legend(["mean" ,"median","Q3","Q1"])
plt.plot(obama.year_month,obama.approve_percent,"o",markersize =2.5,alpha=.15)
 
plt.show()
```
